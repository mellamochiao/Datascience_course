### TASK 1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year


<p>You will require:-
<br>to group the year and calculate the average on the 'Automobile Sales', as the data has years and months column

```python
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum()

plt.figure(figsize=(9, 6))
plt.plot(yearly_sales.index, yearly_sales.values, label='Automobile_Sales')

plt.title('Automobile Sales per Year')
plt.xlabel('Year')
plt.ylabel('Total Yearly Sales')

# Annotating the recessions (1980-1982 and 2000-2001)
plt.annotate('Recession 1980-1982', xy=(1981, 10000), xytext=(1980, 40000),
             arrowprops=dict(color='blue', arrowstyle='->', linewidth=2))
plt.annotate('Recession 2000-2001', xy=(2001, 10000), xytext=(1999, 50000),
             arrowprops=dict(color='blue', arrowstyle='->', linewidth=2))

# Add a legend
plt.legend()

# Rotate the x-axis labels for better readability
years_to_display = [1980, 1981, 1982, 1983, 1990, 1991, 1992, 2000, 2001, 2002, 2008, 2009, 2020]
plt.xticks(years_to_display, rotation = 90)


# Show the plot
plt.tight_layout()
plt.show()
```

###  TASK 2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<p>You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame
    <br>later set year as index and groupby vehicle over Sales and plot
    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title
</p>


```python
# 創建一個只包含經濟衰退期間的 DataFrame
recession_data = df[df['Recession'] == 1]

grouped_recession_df = recession_data.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].sum()
```

```python
vehicle_types = df['Vehicle_Type'].unique()

plt.figure(figsize=(8, 5))

# 為每個車型單獨繪製折線圖
for idx, vehicle in enumerate(vehicle_types):
    # 選擇當前車型的資料
    vehicle_data = grouped_recession_df[grouped_recession_df['Vehicle_Type'] == vehicle]
    
    # 繪製當前車型的折線圖
    plt.plot(vehicle_data['Year'], vehicle_data['Automobile_Sales'], label=vehicle)

# 添加標題和軸標籤
plt.xlabel('Year')
plt.ylabel('Automobile Sales')

plt.legend(title='Vehicle Type')
plt.tight_layout()
plt.show()
```

### TASK 5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<p>You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>


```python
grouped_non_recession = non_recession_data.groupby('Month', as_index=False).agg({
    'Automobile_Sales': 'mean', 
    'Seasonality_Weight': 'mean'
})

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
# 將 'Month' 列轉換為有序的分類類型
grouped_non_recession['Month'] = pd.Categorical(grouped_non_recession['Month'], categories=month_order, ordered=True)
# 根據月份順序排序
grouped_non_recession_sorted = grouped_non_recession.sort_values('Month').reset_index(drop=True)

bubble_size_factor = 100  # Adjust this value to scale bubble size

# Create the bubble plot
plt.figure(figsize=(6, 5))
plt.scatter(grouped_non_recession_sorted['Month'], grouped_non_recession_sorted['Automobile_Sales'], 
            s=grouped_non_recession_sorted['Seasonality_Weight'] * bubble_size_factor,  # Scale the Seasonality Weight for bubble size
            color='blue', alpha=0.6)

plt.title('Seasonality Impact on Automobile Sales', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Automobile Sales', fontsize=12)

legend_sizes = [0.25, 0.5, 0.75, 1.0, 1.25, 1.5]
handles = [plt.scatter([], [], s=size * bubble_size_factor, color='blue', alpha=0.6, label=str(size)) for size in legend_sizes]
plt.legend(handles=handles, title="Seasonality_Weight", loc="lower right")

plt.tight_layout()
plt.show()
```
